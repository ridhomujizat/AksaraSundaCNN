```python
import glob
import numpy as np
import pandas as pd 
import os
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
def prepare_dataset(path):
    x_train=[]
    y_train=[]
    label = 0
    for labels in os.listdir(path):
        if labels == 'ka':
            label = 0
        elif labels == 'ca':
            label = 1
        elif labels == 'ta':
            label = 2
        elif labels == 'pa':
            label = 3
        elif labels == 'ya':
            label = 4
        elif labels == 'wa':
            label = 5
        elif labels == 'ga':
            label = 6
        elif labels == 'ja':
            label = 7
        elif labels == 'da':
            label = 8
        elif labels == 'ba':
            label = 9
        elif labels == 'ra':
            label = 10
        elif labels == 'sa':
            label = 11
        elif labels == 'nga':
            label = 12
        elif labels == 'nya':
            label = 13
        elif labels == 'na':
            label = 14
        elif labels == 'ma':
            label = 15
        elif labels == 'la':
            label = 16
        elif labels== 'ha':
            label = 17
            
        all_images_path=glob.glob(path+labels+'/*.jpg')
        for img_path in all_images_path :
                img=load_img(img_path, target_size=(48,48))
                img=img_to_array(img)
                img=img/255.0
                x_train.append(img)
                y_train.append(label)
    return np.array(x_train),np.array(y_train)
```

```python
x_train, y_train = prepare_dataset('train/')
```

```python
x_test, y_test = prepare_dataset('test/')
```

```python
from keras.preprocessing.image import ImageDataGenerator
import dataaug

datagen = ImageDataGenerator(
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.3,
        horizontal_flip=False,
        preprocessing_function=lambda x: dataaug.elastic_transform(x, alpha_range=8, sigma=3),
        fill_mode='nearest')

datagen.fit(x_train, augment=True)
data_test = datagen.flow(x_train, y_train, batch_size=5)
```

```python
import os
from tensorflow.keras import layers
from tensorflow.keras import Model
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import VGG16

def modelVgg16():
    local_weights_file = 'D:/Belajar/skripsi/ngalegena/pretrained/vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5'
    pretrained_model=VGG16(input_shape = (48, 48, 3), 
                            include_top = False, 
                            weights =None)

    pretrained_model.load_weights(local_weights_file)
    for layer in pretrained_model.layers:
         layer.trainable = False

    x = layers.Flatten()(pretrained_model.output)
    x = layers.Dense(1024, activation='relu')(x)
    x = layers.Dropout(0.5)(x)                  
    x = layers.Dense(18, activation='softmax')(x)           
    model = Model(pretrained_model.input, x) 
    model.compile(optimizer = Adam(lr=0.0001), 
                  loss = 'sparse_categorical_crossentropy', 
                  metrics = ['acc'])
    return model

modelVgg16().summary()
```

```python
model_aug = modelVgg16()
model_without = modelVgg16()
```

```python
history = model_without.fit(x_train,y_train,epochs=20,validation_data=(x_test,y_test))
```

```python
history_Aug = model_aug.fit(data_test,validation_data=(x_test,y_test),
          steps_per_epoch=len(x_train) / 5, epochs=20)
```

```python
import matplotlib.pyplot as plt
from random import randint
import matplotlib.pyplot as plt
plt.plot(history_Aug.history['val_acc'])
plt.plot(history.history['val_acc'])
plt.plot(history_Aug.history['acc'])
plt.plot(history.history['acc'])
plt.title('accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Augmented val', 'Non-Augmeted val','Augmeted train', 'Non-Augmeted train'], loc='upper left')
plt.show()

plt.plot(history_Aug.history['val_loss'])
plt.plot(history.history['val_loss'])
plt.plot(history_Aug.history['loss'])
plt.plot(history.history['loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Augmented val', 'Non-Augmeted val','Augmeted train', 'Non-Augmeted train'], loc='upper left')
plt.show()
```

```python
def pred_dataset(path):
    x_train=[]
    all_images_path=glob.glob(path+'/*.jpg')
    for img_path in all_images_path :
            img=load_img(img_path, target_size=(48,48))
            img=img_to_array(img)
            img=img/255.0
            x_train.append(img)
    return np.array(x_train)
```

```python
pred_img = pred_dataset('pred/aPredic/')
pred_img.shape
```

```python
def get_classlabel(class_code):
    labels = {0:'ka', 1:'ca', 2:'ta', 3:'pa', 4:'ya', 5:'wa', 6:'ga', 7:'ja', 8:'da',
              9:'ba', 10:'ra', 11:'sa', 12:'nga', 13:'nya', 14:'na', 15:'ma', 16:'la', 17:'ha'}
    classl = labels[class_code]
    
    return classl
```

```python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
from random import randint
fig = plt.figure(figsize=(30, 30))
outer = gridspec.GridSpec(5, 5, wspace=0.2, hspace=0.2)

for i in range(25):
    inner = gridspec.GridSpecFromSubplotSpec(2, 1,subplot_spec=outer[i], wspace=0.1, hspace=0.1)
    rnd_number = randint(0,len(pred_img))
    pred_image = np.array([pred_img[rnd_number]])
    x = np.argmax(model_aug.predict(pred_image)[0])
    for j in range(2):
        if (j%2) == 0:
            ax = plt.Subplot(fig, inner[j])
            ax.imshow((pred_image[0]*255).astype(np.uint8))
            ax.set_title(get_classlabel(x))
            ax.set_xticks([])
            ax.set_yticks([])
            fig.add_subplot(ax)


fig.show()
```
